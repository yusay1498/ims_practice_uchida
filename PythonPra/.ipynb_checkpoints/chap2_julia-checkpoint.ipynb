```julia
using DataFrames
using CSV
using Plots
using Random
using LinearAlgebra

# Perceptron class
struct Perceptron
    eta::Float64
    n_iter::Int
    w::Vector{Float64}
    errors::Vector{Int}
    random_state::Int
end

function Perceptron(; eta=0.01, n_iter=50, random_state=1)
    w = randn(1 + 2) * 0.01  # 2 features for Iris dataset
    return Perceptron(eta, n_iter, w, Int[], random_state)
end

function fit!(p::Perceptron, X::Matrix{Float64}, y::Vector{Int})
    Random.seed!(p.random_state)
    for _ in 1:p.n_iter
        errors = 0
        for i in 1:size(X, 1)
            update = p.eta * (y[i] - predict(p, X[i, :]))
            p.w[2:end] .+= update * X[i, :]
            p.w[1] += update
            errors += update != 0.0 ? 1 : 0
        end
        push!(p.errors, errors)
    end
    return p
end

function net_input(p::Perceptron, X::Vector{Float64})
    return dot(X, p.w[2:end]) + p.w[1]
end

function predict(p::Perceptron, X::Vector{Float64})
    return net_input(p, X) >= 0.0 ? 1 : -1
end

# Load the Iris dataset
df = CSV.read("/home/ubuntu/repositories/ims_practice_uchida/PythonPra/iris.data", DataFrame, header=false)
y = df[1:100, 5] .== "Iris-setosa" # Get first 100 entries
y = Int.(y) .+ 1  # Convert Boolean to 1 and 2 (1 for setosa, 2 for versicolor)

# Convert specific columns to Matrix
X = hcat(Float64.(df[1:100, 1]), Float64.(df[1:100, 3]))  # Sepal length and petal length

# Plotting
scatter(X[1:50, 1], X[1:50, 2], color=:red, label="setosa", legend=:topright)
scatter!(X[51:100, 1], X[51:100, 2], color=:blue, label="versicolor")
xlabel!("sepal length [cm]")
ylabel!("petal length [cm]")
title!("Iris Dataset")

# Train the Perceptron
ppn = Perceptron(eta=0.1, n_iter=10)
fit!(ppn, X, y)  # y is now Vector{Int}

# Plotting error updates
plot(1:length(ppn.errors), ppn.errors, marker=:o, xlabel="Epochs", ylabel="Number of updates", title="Perceptron Training")

# Decision region plotting function
function plot_decision_regions(X, y, classifier)
    x_min, x_max = minimum(X[:, 1]) - 1, maximum(X[:, 1]) + 1
    y_min, y_max = minimum(X[:, 2]) - 1, maximum(X[:, 2]) + 1
    xx1 = range(x_min, stop=x_max, length=200)
    xx2 = range(y_min, stop=y_max, length=200)
    
    Z = [predict(classifier, [x1, x2]) for x1 in xx1, x2 in xx2]
    Z = reshape(Z, length(xx1), length(xx2))

    contourf(xx1, xx2, Z, alpha=0.3)
    scatter!(X[:, 1], X[:, 2], color=(y .== 1) .+ 1, label="samples")
end

plot_decision_regions(X, y, ppn)
xlabel!("sepal length [cm]")
ylabel!("petal length [cm]")
title!("Decision Regions")
```
